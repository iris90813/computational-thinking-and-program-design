```
#BMI計算

##輸入體重和身高資訊input
weight = float(input("請輸入你的體重(KG)? "))
height = float(input("請輸入你的身高(M)?" ))

##計算BMI指數process
bmi = weight / (height * height)

##印出BMI output
print("BMI指数為",bmi)

##根據BMI指数印出肥胖程度分類 process + ouput + if elif else
if bmi< 18 :
    print("過瘦")
elif bmi < 24:
    print("標準")
elif bmi <27:
    print("過重")
else:
    print("肥胖")





```
